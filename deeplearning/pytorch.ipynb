```python
#perceptron


import torch
import matplotlib.pyplot as plt
import numpy as np


N = 5

x1 = torch.rand(N)
x2 = x1 + torch.randint(1, 10, [N]) / 10
c1 = torch.vstack([x1, x2]).mT

x1 = torch.rand(N)
x2 = x1 - torch.randint(1, 10, [N]) / 10
c2 = torch.vstack([x1, x2]).mT

f = [0, 1]  # splitting line

w = torch.FloatTensor([-0.3, 0.3])
for i in range(N):
  x = c1[:][i]
  y = torch.dot(w, x)
  if y >= 0:
    print("Class C1")
  else:
    print("Class C2")

plt.scatter(c1[:, 0], c1[:, 1], s=10, c='red')
plt.scatter(c2[:, 0], c2[:, 1], s=10, c='blue')
plt.plot(f)
plt.grid()
plt.show()
```

```python
#perceptron 2 lines splitting


import torch
import matplotlib.pyplot as plt
import numpy as np


N = 5
b = 3

x1 = torch.rand(N)
x2 = x1 + torch.randint(1, 10, [N]) / 10 + b
c1 = torch.vstack([x1, x2, torch.ones(N)]).mT

x1 = torch.rand(N)
x2 = x1 - torch.randint(1, 10, [N]) / 10 + b
c2 = torch.vstack([x1, x2, torch.ones(N)]).mT

f = [0 + b, 1  + b]  # splitting line
w1 = -0.5
w2 = -w1
w3 = -b * w2
w = torch.FloatTensor([w1, w2, w3])

for i in range(N):
  x = c1[:][i]
  y = torch.dot(w, x)
  if y >= 0:
    print("Class C1")
  else:
    print("Class C2")

plt.scatter(c1[:, 0], c1[:, 1], s=10, c='red')
plt.scatter(c2[:, 0], c2[:, 1], s=10, c='blue')
plt.plot(f)
plt.grid()
plt.show()
```
